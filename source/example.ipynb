# New Projects

The first step is to import the package and create a MarCon dictionary. This is one giant dictionary that contains all required options, filepaths, etc. These can also be produced using the GUI [Marxan Connect](marxanconnect.ca)

```python
import marxanconpy
import os

project = marxanconpy.marcon.new_project()

project
```

We can also load pre-existing `.Marcon` files

```python
path = os.path.dirname(marxanconpy.__file__)
project = marxanconpy.marcon.load_project(os.path.join(path,'data','CF_demographic','tutorial.MarCon'))
project
```
